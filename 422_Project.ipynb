# **Pre-process for KNN**

**Scaling & Encoding**

```python
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_features)
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
categorical_features
```

**Correlation Analysis**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df_corr = df.copy()

category_fe = df_corr.select_dtypes(include=['object']).columns.drop('NObeyesdad')
df_corr = pd.get_dummies(df_corr, columns=category_fe, drop_first=True)

df_corr['NObeyesdad'] = LabelEncoder().fit_transform(df_corr['NObeyesdad'])

corr_matrix = df_corr.corr()

plt.figure(figsize=(18,15))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (All Features vs Target)")
plt.show()

target_corr = corr_matrix['NObeyesdad'].sort_values(ascending=False)
print("Feature correlation with target:\n", target_corr)
```

**Train-test splitting**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x=df.drop('NObeyesdad',axis=1)
y=df['NObeyesdad']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
categorical_features=x.select_dtypes(include=['object']).columns
numerical_features=x.select_dtypes(include=['int64', 'float64']).columns
```

```python
scaler=StandardScaler()

x_train[numerical_features] = scaler.fit_transform(x_train[numerical_features])
x_test[numerical_features]= scaler.transform(x_test[numerical_features])
```

```python
print('x_train after scaling:')
display(x_train[numerical_features].head())
```

```python
print('Mean of numerical features in x_train after scaling:')
display(x_train[numerical_features].mean())
print('Standard deviation of numerical features in x_train after scaling:')
display(x_train[numerical_features].std())
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd

le_y = LabelEncoder()
le_y.fit(pd.concat([y_train, y_test]))

y_train = le_y.transform(y_train)
y_test = le_y.transform(y_test)

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

x_train_categorical_encoded = ohe.fit_transform(x_train[categorical_features])
x_test_categorical_encoded = ohe.transform(x_test[categorical_features])

ohe_feature_names = ohe.get_feature_names_out(categorical_features)

x_train_categorical_df = pd.DataFrame(x_train_categorical_encoded,columns=ohe_feature_names,index=x_train.index)

x_test_categorical_df = pd.DataFrame(x_test_categorical_encoded,columns=ohe_feature_names,index=x_test.index)

x_train = x_train.drop(columns=categorical_features)
x_test = x_test.drop(columns=categorical_features)

x_train = pd.concat([x_train, x_train_categorical_df], axis=1)
x_test = pd.concat([x_test, x_test_categorical_df], axis=1)
```

```python
print('x_train after all transformations:')
display(x_train.head())
print('\nx_test after all transformations:')
display(x_test.head())
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train, y_train)
```

```python
knn_y_pred = knn_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score as knn_accuracy_score, classification_report as knn_classification_report,confusion_matrix as knn_confusion_matrix

print("Accuracy:", knn_accuracy_score(y_test, knn_y_pred))
print("Classification Report:\n", knn_classification_report(y_test, knn_y_pred))
print("Confusion Matrix:\n", knn_confusion_matrix(y_test, knn_y_pred))
```

## **Naive Bayes**

## **Data Preparation for Naive Bayes**

```python
x_nb = df.drop('NObeyesdad', axis=1)
y_nb = df['NObeyesdad']

x_train_nb, x_test_nb, y_train_nb, y_test_nb = train_test_split(x_nb, y_nb, test_size=0.2, random_state=23)



scaler_nb = StandardScaler()
x_train_nb[numerical_features] = scaler_nb.fit_transform(x_train_nb[numerical_features])
x_test_nb[numerical_features] = scaler_nb.transform(x_test_nb[numerical_features])

le_y_nb = LabelEncoder()
y_train_nb_encoded = le_y_nb.fit_transform(y_train_nb)
y_test_nb_encoded = le_y_nb.transform(y_test_nb)


for col in categorical_features:
    le_feature = LabelEncoder()
    all_categories = pd.concat([x_train_nb[col], x_test_nb[col]]).astype(str).unique()
    le_feature.fit(all_categories)

    x_train_nb[col] = le_feature.transform(x_train_nb[col])
    x_test_nb[col] = le_feature.transform(x_test_nb[col])

print('x_train_nb after all transformations:')
display(x_train_nb.head())
print('\nx_test_nb after all transformations:')
display(x_test_nb.head())
```

```python
print('x_train after all transformations:')
display(x_train.head())
print('\nx_test after all transformations:')
display(x_test.head())
```

# **NAIVE BAYES**

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()
nb_model.fit(x_train_nb, y_train_nb_encoded)
```

```python
nb_y_pred = nb_model.predict(x_test_nb)
```

```python
from sklearn.metrics import accuracy_score as nv_accuracy_score, classification_report as nv_classification_report, confusion_matrix as nv_confusion_matrix

print("Accuracy:", nv_accuracy_score(y_test_nb_encoded, nb_y_pred))
print("\nClassification Report:", nv_classification_report(y_test_nb_encoded, nb_y_pred))
print("\nConfusion Matrix:", nv_confusion_matrix(y_test_nb_encoded, nb_y_pred))
```

## **Report**

```python
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import numpy as np

y_prob = nb_model.predict_proba(x_test_nb)

classes = np.unique(y_test_nb_encoded)
y_test_bin = label_binarize(y_test_nb_encoded, classes=classes)

fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f'Naive Bayes (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve – Naive Bayes (Multiclass)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

knn_accuracy = accuracy_score(y_test, knn_y_pred)

nb_accuracy = accuracy_score(y_test_nb_encoded, nb_y_pred)

models = ['KNN', 'Naive Bayes']
accuracies = [knn_accuracy, nb_accuracy]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['skyblue', 'lightcoral'])
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Prediction Accuracy of Models')
plt.ylim(0, 1)

for i, v in enumerate(accuracies):
    plt.text(i, v + 0.01, f"{v:.2f}", ha='center')

plt.show()
```
